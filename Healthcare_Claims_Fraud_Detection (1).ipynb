```python
import pandas as pd

path = "/content/drive/MyDrive/Medicare_Physician_Other_Practitioners_by_Provider_and_Service_2024.csv"

chunks = []
chunk_size = 100000

for chunk in pd.read_csv(path, chunksize=chunk_size):
    chunks.append(chunk.sample(frac=0.05, random_state=42))

sample = pd.concat(chunks)

print(sample.shape)

sample.to_csv("cms_sample.csv", index=False)
```
